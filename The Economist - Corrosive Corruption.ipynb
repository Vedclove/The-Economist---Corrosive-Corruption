```r
df <- read.csv("Economist_Assignment_Data.csv")
```

```r
library(ggplot2)
```

```r
print(head(df))
```

```r
# Use ggplot() + geom_point() to create a scatter plot object called pl. You will need to specify x=CPI and y=HDI and color=Region as aesthetics
```

```r
pl2 <- ggplot(df , aes(x = CPI , y = HDI))
```

```r
pl <- pl2 + geom_point(size = 2 , aes(color = Region))
```

```r
print(pl)
```

```r
# Change the points to be larger empty circles. (You'll have to go back and add arguments to geom_point() and reassign it to pl.) You'll need to figure out what shape= and size=
```

```r
new.pl <- pl2 + geom_point(size = 5 , shape = 1 , aes(color = Region))
```

```r
new.pl
```

```r
# Add a trend line to the data
```

```r
trend.pl <- new.pl + geom_smooth()
```

```r
trend.pl
```

```r
trend.pl <- new.pl + geom_smooth(method = 'lm' , formula = y~log(x) , se = FALSE , color = 'red')
```

```r
trend.pl
```

```r
# Adding the label for the countries
```

```r
text.country <- trend.pl + geom_text(aes(label = Country , color = Region))
```

```r
text.country
```

```r
# Using specific country values
```

```r
pointsToLabel <- c("Russia", "Venezuela", "Iraq", "Myanmar", "Sudan",
                   "Afghanistan", "Congo", "Greece", "Argentina", "Brazil",
                   "India", "Italy", "China", "South Africa", "Spane",
                   "Botswana", "Cape Verde", "Bhutan", "Rwanda", "France",
                   "United States", "Germany", "Britain", "Barbados", "Norway", "Japan",
                   "New Zealand", "Singapore")
```

```r
pl3 <- trend.pl + geom_text(aes(label = Country), color = "gray20", 
                data = subset(df, Country %in% pointsToLabel),check_overlap = TRUE)
```

```r
pl3
```

```r
# Beautification
```

```r
pl4 <- pl3 + theme_bw() + scale_x_continuous(name = "Corruption Perceptions Index, 2011 (10 = least corrupt)" , breaks = 0:10)
```

```r
pl4
```

```r
pl5 <- pl4 + scale_y_continuous(name = "Human Development Index, 2011 (1=best)") + ggtitle()
```
